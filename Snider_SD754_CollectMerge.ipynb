**Collect and Merge Data from Two NBA Datasets**

**Import Two Datasets into Data Frames**

```python
import pandas as pd

statistics_df = pd.read_csv('Statistics.csv', encoding='ISO-8859-1')
print("\nSample set for Statistics dataset:")
print(statistics_df.head())
```

```python
salaries_df = pd.read_csv('Salaries.csv')
print("Sample set for Salaries dataset:")
print(salaries_df.head())
```

**Information about the Datasets**

```python
print("\nData types for the Statistics dataset:")
print(statistics_df.dtypes)
```

```python
print("\nData types for the Salaries dataset:")
print(salaries_df.dtypes)
```

```python
print("\nSummary statistics for the Statistics dataset:")
print(statistics_df.describe())
```

```python
print("Summary statistics for the Salaries dataset:")
print(salaries_df.describe())
```

**Create Subsets of Each Dataset**

```python
statistics_subset = statistics_df[['Player', 'Team', 'Pos', 'Age']]
statistics_subset
```

```python
salaries_subset = salaries_df[['Player', '2024-25']]
salaries_subset
```

**Inner Merge of the Two Subsets**

```python
# Clean the '2024-25' salary column by removing "$" and commas to convert it into numeric values
salaries_df['2024-25'] = salaries_df['2024-25'].replace({r'\$': '', r',': ''}, regex=True).astype(float)

# Inner merge the Salaries subset with the Statistics subset on 'Player'
# An inner merge is being performed because I want to only keep rows with matching keys in both frames
merged_df = pd.merge(salaries_subset, statistics_subset, on='Player', how='inner')
merged_df
```
