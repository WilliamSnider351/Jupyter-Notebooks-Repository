```python
import pandas as pd
```

**Reading Summits excel sheet from the country.xlsx file into a pandas dataframe.**

```python
summits = pd.read_excel("country.xlsx", sheet_name = "Summits")
summits
```

**Overview with the info method.**

```python
summits.info()
```

**Using the describe function to display information about the leader feature in the summits dataframe.**

```python
summits[['leader']].describe()
```

**Using the value_count function to sum and sort by leader.**

```python
summits[['leader']].value_counts()
```

**Reading the vehicle csv file into a dataframe.**

```python
vehicles = pd.read_csv("vehicles.csv")
vehicles
```

**Overview with the info method.**

```python
vehicles.info()
```

**Using the head function to display the first 5 rows of the dataframe.**

```python
vehicles.head()
```

**Using the describe function to display summary information about the co2emissions feature.**

```python
vehicles['co2emissions'].describe()
```

**Aggregating the model and emissions features, grouping by model, finding the mean emissions and sorting by emissions.**

```python
vehicles.groupby('model')[['co2emissions']].mean().sort_values(by = 'co2emissions')
```

**Aggregating by make and cylinders features, grouping by make, finding the sum, min and max of cylinders.**

```python
vehicles.groupby('make')[['cylinders']].agg(['sum', 'min', 'max'])
```
